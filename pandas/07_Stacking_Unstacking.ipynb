# Stacking and Unstacking

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A': ['one', 'one', 'two', 'three'] * 3,
                   'B': ['male', 'female', 'intersex'] * 4,
                   'C': ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 2,
                   'D': np.random.randint(6, size=12),
                   'E': np.random.randint(4, size=12)})
```

```python
df
```

```python
df.set_index(['A', 'B', 'C'], inplace=True)
df
```

```python
# stack()
df_stack = df.stack()
df_stack
```

```python
# unstack
df_unstack = df_stack.unstack() 
df_unstack
```

```python
# unstack by index
df_unstack_A = df_stack.unstack(0) #or 'A' 
df_unstack_A
```
