# Pivot Tables

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A': ['one', 'one', 'two', 'three'] * 3,
                    'B': ['male', 'female', 'intersex'] * 4,
                    'C': ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 2,
                    'D': np.random.randint(6, size=12),
                    'E': np.random.randint(4, size=12)})
```

```python
df
```

```python
pd.pivot_table(df, values='D', index=['A'], columns=['C'], aggfunc=np.sum) # default aggfunc = np.mean
```

```python
# if columns are not given
pd.pivot_table(df, values='D', index=['A', 'B'], aggfunc=np.sum) 
```

```python
# if valuebbs are not given, it will return all values
pd.pivot_table(df, index=['A'], columns=['C'], aggfunc=np.sum) 
```

```python
# adding margins, which can insert one column "ALL" to the pivot table
df.pivot_table(values='D', index=['A', 'B'], columns=['C'], aggfunc=np.sum, margins=True)
```
